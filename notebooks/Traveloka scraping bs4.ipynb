## Multiple Page Scraping

```python
import requests
import pandas as pd
import re

from bs4 import BeautifulSoup
pd.set_option('display.max_columns',None)

#global dataframe
data_combined = pd.DataFrame()

for page in range (1,30):
  print (f'Processing page {page}')
  if page==1:
    url = 'https://www.traveloka.com/en-en/accommodation/villa/indonesia/region/bali-102746'
  else:
    url = f'https://www.traveloka.com/en-en/accommodation/villa/indonesia/region/bali-102746/{page}'
  response = requests.get(url)
  soup = BeautifulSoup(response.text,'html.parser')

  containers = soup.find_all('div',attrs={'data-testid':'popular-hotel-card'})

  #list all villa names
  villa_name = []
 
  for name in containers:
    try:
      name.find('div',attrs={'class':'css-1dbjc4n r-13awgt0 r-18u37iz'})
      villa_name.append(name.find('h3',attrs={'data-testid':'popular-hotel-card-name'}).text)
    except:
      villa_name.append('')

  #list all features
  features = []
  
  for feat in containers:
    try:
      feature_divs = feat.find_all('div', attrs={'class': 'css-1dbjc4n r-1awozwy r-sdzlij r-1kbdv8c r-18u37iz r-s1qlax r-1vvnge1'})
      #extract test from each feature div
      feature_texts = [div.text.strip() for div in feature_divs if div.text.strip()]
      features.append(', '.join(feature_texts) if feature_texts else '')
    except:
      features.append('')

  #list all prices
  price = []
  
  for usd in containers:
    try:
      usd.find_all('div',attrs={'class':'css-1dbjc4n r-3mtglp r-13qz1uu'})
      price.append(usd.find('h2',attrs={'data-testid':'popular-hotel-price'}).text)
    except:
      price.append('')

  #list all rating
  rating = []
  
  for rate in containers:
    try:
      rate.find_all('div',attrs={'class':'css-1dbjc4n r-1awozwy r-18u37iz r-1h0z5md'})
      rating.append(rate.find('div',attrs={'data-testid':'popular-hotel-rating'}).text)
    except:
      rating.append('')

  #list all total reviews
  total_review = []
  
  for tot_rev in containers:
    try:
      tot_rev.find_all('div',attrs={'class':'css-1dbjc4n r-1awozwy r-18u37iz r-1h0z5md'})
      total_review.append(tot_rev.find('div',attrs={'data-testid':'popular-hotel-total-review'}).text)
    except:
      total_review.append('')

  #list all rating info  
  rating_info = []
  
  for info in containers:
    try:
      info.find_all('div',attrs={'class':'css-1dbjc4n r-1awozwy r-18u37iz r-1h0z5md'})
      rating_info.append(info.find('div',attrs={'data-testid':'popular-hotel-rating-info'}).text)
    except:
      rating_info.append('')

  #list all locations
  location = []
  
  for loc in containers:
    try:
      loc.find_all('div',attrs={'class':'css-1dbjc4n r-1awozwy r-18u37iz r-1h0z5md'})
      location.append(loc.find('div',attrs={'data-testid':'popular-hotel-address'}).text)
    except:
      location.append('')

  #list all URL
  url = []
  
  for link in containers:
    try:
      link.find_all('div',attrs={'class':'css-1dbjc4n r-hv712p'})
      url.append('www.traveloka.com'+link.find('a')['href'])
    except:
      url.append('')
    
  df = pd.DataFrame ({
    'VillaName': villa_name,
    'Features': features,
    'Price': price,
    'Rating': rating,
    'Total_Review': total_review,
    'Rating_Info': rating_info,
    'Location': location,
    'URL': url
    })

  data_combined = pd.concat([data_combined,df])
  print(f'End of page {page}')
```

## Data Cleaning

```python
traveloka_data = data_combined.reset_index(drop = True)
traveloka_data
```

```python
#make price uniform
traveloka_data['Currency'] = traveloka_data['Price'].apply(lambda x:str(x).split(' ')[0])
traveloka_data['PriceValue'] = traveloka_data['Price'].apply(lambda x:str(x).split(' ')[1])
traveloka_data[['Price','Currency','PriceValue']]
```

```python
#check
traveloka_data['Currency'].unique()
```

```python
#fix price
traveloka_data['PriceValue'] = traveloka_data['PriceValue'].astype(str).str.replace(',', '').astype(float)
```

```python
#drop redundant
traveloka_data = traveloka_data.drop(columns='Price', axis = 1)
```

```python
#make rating uniform and numerical
traveloka_data['Rating'] = pd.to_numeric(traveloka_data['Rating'])
```

```python
#remove parantheses in total review for standardization
traveloka_data['Total_Review'] = traveloka_data['Total_Review'].apply(lambda x:str(x).replace('(','').replace(')','').replace(',',''))
traveloka_data['Total_Review'] = pd.to_numeric(traveloka_data['Total_Review'],errors ='coerce')
```

```python
#extract specific area
#clean address to form a new area column
target_locations =['Canggu', 'Umalas', 'Seminyak', 'Legian', 'Uluwatu', 'Jimbaran', 'Uluwatu', 'Pecatu', 'Kutuh',
                   'Ubud', 'Sanur', 'Nusa Dua', 'Ungasan', 'Lembongan', 'Nusapenida', 'Denpasar']

def extract_location(row, locations):
    #combine VillaName and Address into one string for checking
    combined_text = str(row['Location'])
    
    #iterate through the locations list and check if any of the locations are in the combined text
    for location in locations:
        if location.lower() in combined_text.lower():
            return location
    return 'Others'  #if no match found
```

```python
traveloka_data['Area'] = traveloka_data.apply(lambda row: extract_location(row, target_locations), axis=1)
```

```python
#check missing data
import numpy as np

df = traveloka_data.copy()
df['Rating_Info'] = df['Rating_Info'].replace('', np.nan)
df['Features'] = df['Features'].replace('', np.nan)


missing = df.isnull().sum().reset_index()
missing.columns = ['features', 'missing_values']
missing['percentage'] = (missing['missing_values']/len(df)) *100
missing
```

```python
#fill
df['Rating'] = df['Rating'].fillna(df['Rating'].median())
df['Total_Review'] = df['Total_Review'].fillna(df['Total_Review'].median())
df['Rating_Info'] = df['Rating_Info'].fillna(df['Rating_Info'].mode()[0])
```
